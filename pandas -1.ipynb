```python
# 1.Find out how may male and female passengers was onboarded   

print("no of male passenger onboard :- ",len(c[c["Sex"]=="male"]))
print("no of female passenger onboard :- ",len(c[c["Sex"]=="female"]))

# or

c["Sex"].value_counts()

# or

c.groupby("Sex")["Sex"].count()
```

```python
# 2.how many servived we have

print("people servived :- ",len(c[c["Survived"]==1]))

# or

c["Survived"].value_counts()
```

```python
# 3.how many casulity we have

len(c[c["Survived"]==0])
```

```python
# 4.what is name of a person who is the eldest one

c[c["Age"]==max(c["Age"])]["Name"]
```

```python
# 5.how many passengers do we have in first , second and third class

print("passengers in first class :- ",len(c[c["Pclass"]==1]))
print("passengers in second class :- ",len(c[c["Pclass"]==2]))
print("passengers in thirst class :- ",len(c[c["Pclass"]==3]))

# or

c["Pclass"].value_counts()

```

```python
# 6.how many person we have whoes name starts with "s"

c.loc[(c["Name"].str.startswith("S"))]

# or 

#c[(c["Name"].str.startswith("S"))]
```

```python
# 8.how many person do we have below age of 25

print("person having age is less than 25 :=", len(c[c["Age"]<25]))
```

```python
# 9.how many person died whose age was less than 40

print("person whose died and Age is less than 40 :- ",len(c.loc[(c["Age"]<40) & (c["Survived"]==0)]))
c.loc[(c["Age"]<40) & (c["Survived"]==0)].head()
```
